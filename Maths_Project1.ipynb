```python
# convert data to binary classification rating>=4 is 0 rating==5 is 1

df['Rating'] = df['Rating'].replace(1, 0)
df['Rating'] = df['Rating'].replace(2, 0)
df['Rating'] = df['Rating'].replace(3, 0)
df['Rating'] = df['Rating'].replace(4, 0)
df['Rating'] = df['Rating'].replace(5, 1)
df['Rating']
```

```python
# Defining X as list of Reviews

X=list(df['Review'])
```

```python
# Defining y as Ratings
y=list(df['Rating'])
```

```python
# Spliting the data into train and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# Installing the transformers

%pip install transformers
```

```python
# Initiating the Distil Bert Transformer

from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
# Encoding

train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
y_train
```

```python
# Train and test encoding

import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
from transformers import TFDistilBertForSequenceClassification, TFTrainer, TFTrainingArguments

training_args = TFTrainingArguments(
    output_dir='./finalresult',          # output directory
    num_train_epochs=5,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)
```

```python
training_args.eval_steps = 1000
```

```python
# Training the model

with training_args.strategy.scope():
    model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = TFTrainer(
    model=model,                         
    args=training_args,                  
    train_dataset=train_dataset,         
    eval_dataset=test_dataset             
)
# trainer.train()
```

```python
# Evaluating the model and displaying the loss

trainer.evaluate(test_dataset)
```

```python
# Predicting of test dataset

trainer.predict(test_dataset)
```
